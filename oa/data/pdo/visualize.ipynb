```python
from ase import Atoms
from ase.visualize.plot import plot_atoms
import py3Dmol
import io
import numpy as np

data = np.load('train.npz', allow_pickle=True)
mol = data['FS'][0]
```

```python
atoms = Atoms(
    numbers=mol[:, 0].astype(int),
    positions=mol[:, 1:4],
)
xyz_stringio = io.StringIO()
atoms.write(xyz_stringio, format='xyz')
xyz_string = xyz_stringio.getvalue()

view = py3Dmol.view(width=800, height=400)
view.addModel(xyz_string, 'xyz')
view.setStyle({'sphere': {'scale': 0.5}, 'stick': {'radius': 0.1}})
view.zoomTo()
view.show()
```
